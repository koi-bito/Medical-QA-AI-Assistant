# 01 — Dataset Exploration

**Goal:** Download the `lavita/ChatDoctor-HealthCareMagic-100k` dataset and understand its structure.

**Kernel:** Make sure you're using the `Medical QA` kernel (top-right in VS Code).

```python
from datasets import load_dataset

# Load the dataset (downloads ~200MB on first run)
print('Loading dataset...')
dataset = load_dataset('lavita/ChatDoctor-HealthCareMagic-100k')

print('Dataset structure:', dataset)
print(f'\nTotal training examples: {len(dataset["train"])}')
print('\n--- First Example ---')
print(dataset['train'][0])
```

```python
import pandas as pd

df = pd.DataFrame(dataset['train'])
print('Columns:', df.columns.tolist())
print('\nData types:')
print(df.dtypes)
print('\nNull counts:')
print(df.isnull().sum())
```

```python
print('\n--- 5 Random Examples ---\n')
sample = df.sample(5, random_state=42)
for _, row in sample.iterrows():
    print('INSTRUCTION:', row['instruction'][:100])
    print('INPUT:', row['input'][:200])
    print('OUTPUT:', row['output'][:300])
    print('-' * 80)
```
